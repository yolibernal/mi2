# Inverse cdf method

## Inverse Laplace cdf

## Samples from Laplace distribution

```python
import numpy as np

num_samples = 10000
samples = np.random.uniform(0, 1, num_samples)
```

```python
mu = 1
b = 2
def F_Laplace_inv(z, mu=mu, b=b):
    if z < 0.5:
        return b * np.log(2 * z) + mu
    if z >= 0.5:
        return -b * np.log(-2 * z + 2) + mu
```

```python
F_Laplace_inv(0.5)
```

```python
samples_laplace = [F_Laplace_inv(z) for z in samples]
```

## Plot density estimate (normalized histogram)

```python
import matplotlib.pyplot as plt
from scipy.stats import laplace

fig, ax = plt.subplots()
x = np.linspace(-20, 20, 100)
y = laplace.pdf(x, loc=mu, scale=b)
ax.plot(x, y)
ax.hist(samples_laplace, density=True, bins=50)
```

## Using inverse cdf from scipy

`ppf` (percent-point function) is inverse of cdf

```python
fig, ax = plt.subplots()
x = np.linspace(-20, 20, 100)
y = laplace.pdf(x, loc=mu, scale=b)
ax.plot(x, y)
ax.hist(laplace.ppf(samples, loc=mu, scale=b), density=True, bins=50)
```

## Self stuff

```python
from scipy.stats import norm

fig, ax = plt.subplots()

y = norm(loc=0, scale=1).cdf(np.random.normal(size=100000))

ax.hist(y, bins=100)
```
