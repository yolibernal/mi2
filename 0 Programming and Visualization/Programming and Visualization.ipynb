### Manually calculate convariance matrix

Manually calculate (almost) covariance matrix using numpy arrays

```python
csv_centered_matrix = csv_centered.to_numpy()
np.matmul(csv_centered_matrix.transpose(), csv_centered_matrix)
```

Manually calculate covariance matrix using pandas

```python
manual_cov_matrix = csv_centered.transpose().dot(csv_centered) / len(csv_centered.index)
```

### Calculate covariance matrix using built-in functions

```python
np.cov(csv.to_numpy(), rowvar=False)
```

```python
csv.cov()
```

### Compare manually calculated and built-in methods

```python
def draw_cov_heatmap(cov_matrix):
    fig, ax = plt.subplots(figsize=(18, 16))
    im = ax.imshow(cov_matrix)
    size = cov_matrix.shape[0]
    for i in np.arange(size):
        for j in np.arange(size):
            text = ax.text(j, i, "%.2f" % cov_matrix.iloc[i][j],
                           ha="center", va="center", color="w")
```

```python
draw_cov_heatmap(csv.cov())
```

```python
draw_cov_heatmap(manual_cov_matrix)
```

### Rotate unit vector
(see [this video](https://www.youtube.com/watch?v=DOMg0lXWatM))

So the resulting vector is


```python
import math

def rotate_vector(x, theta):
    rotated_x0 = math.cos(theta) * x[0] - math.sin(theta) * x[1]
    rotated_x1 = math.sin(theta) * x[0] + math.cos(theta) * x[1]
    rotated_x = pd.Series([rotated_x0, rotated_x1], index=["x", "y"])
    return rotated_x
```

```python
rotate_vector((1,0), math.pi)
```

```python
rotation_radians = [n / 12 *  math.pi for n in np.arange(12 + 1)]
rotation_radians
```

```python
from functools import partial

rotate_vector_by_90_deg = partial(rotate_vector, theta=math.pi/2)
```

```python
rotate_vector_by_90_deg((1,0))
```

```python
df_xy_rotated_90_deg = df[["x", "y"]].apply(rotate_vector_by_90_deg, result_type="expand", axis=1)
df_xy_rotated_90_deg.head()
```

```python
 df_xy_rotated_90_deg.plot(style=".", x="x", y="y")
```

```python
covariance_matrices = []
for radian in rotation_radians:
    rotate_vector_by_radian = partial(rotate_vector, theta=radian)
    df_xy_rotated = df[["x", "y"]].apply(rotate_vector_by_radian, result_type="expand", axis=1)
    df_xy_rotated.plot(style=".", x="x", y="y")
    covariance_matrices.append(df_xy_rotated.cov())
```

```python
for cov_matrix in covariance_matrices:
    print(f"Covariance matrix for rotation of {radian}")
    print(cov_matrix)
    print()
```

```python
variances_x = [cov_matrix["x"]["x"] for cov_matrix in covariance_matrices]
variances_y = [cov_matrix["y"]["y"] for cov_matrix in covariance_matrices]

plt.plot(rotation_radians, variances_x)
plt.plot(rotation_radians, variances_y)
```

## Exercise 0.8: Data Processing: Image data

```python
from PIL import Image

image = Image.open('natIMG.jpg')
image
```

```python
data = np.asarray(image)
data
```

```python
fig, ax = plt.subplots(figsize=(20, 20))
im = ax.imshow(data, cmap="gray")
```

```python
from sklearn.feature_extraction.image import extract_patches_2d

patches = extract_patches_2d(data, (10, 10), max_patches=100)
```

```python
patches_2d = patches.reshape(10, 10, 10, 10)
assert(all((patches[0] == patches_2d[0][0]).flatten()))
```

```python
fig, axs = plt.subplots(nrows=10, ncols=10, figsize=(20, 20))
for i in np.arange(axs.shape[0]):
    for j in np.arange(axs.shape[1]):
        im = axs[i][j].imshow(patches_2d[i][j], cmap="gray")
```
