```python
from sklearn.manifold import LocallyLinearEmbedding
import numpy as np
```

## Spiral

```python
X_spiral = np.loadtxt("spiral.csv", skiprows=1, usecols=(1,2,3), delimiter=",")
np.random.shuffle(X_spiral)
```

```python
embedding_spiral = LocallyLinearEmbedding(n_components=1, n_neighbors=20)
```

```python
X_spiral_transformed = embedding_spiral.fit_transform(X_spiral)
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(15, 5))

p = X_spiral_transformed.shape[0]
c = X_spiral_transformed[:,0]

ax.scatter(X_spiral_transformed[:,0], np.zeros(p), cmap="rainbow", c=c)

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

plot_3d = plt.figure(figsize=(20,10)).gca(projection='3d')
plot_3d.scatter(X_spiral[:,0], X_spiral[:,1], X_spiral[:,2], cmap="rainbow", c=c)
plot_3d.set_xlabel('x')
plot_3d.set_ylabel('y')
plot_3d.set_zlabel('z')
```

## Bowl

```python
X_bowl = np.loadtxt("bowl.csv", skiprows=1, usecols=(1,2,3), delimiter=",")
embedding_bowl = LocallyLinearEmbedding(n_components=2, n_neighbors=10)

X_bowl_transformed = embedding_bowl.fit_transform(X_bowl)
```

```python
fig, ax = plt.subplots(figsize=(15, 10))

p = X_bowl_transformed.shape[0]
c = X_bowl_transformed[:,0]**0.5

ax.scatter(X_bowl_transformed[:,0], X_bowl_transformed[:,1], cmap="rainbow", c=c)

plot_3d = plt.figure(figsize=(20,10)).gca(projection='3d')
plot_3d.scatter(X_bowl[:,0], X_bowl[:,1], X_bowl[:,2], cmap="rainbow", c=c)
plot_3d.set_xlabel('x')
plot_3d.set_ylabel('y')
plot_3d.set_zlabel('z')
```
